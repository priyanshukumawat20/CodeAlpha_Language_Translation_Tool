```python
import gradio as gr
from googletrans import Translator

translator = Translator()

def translate_text(text, source_lang, target_lang):
    result = translator.translate(text, src=source_lang, dest=target_lang)
    return result.text

demo = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Enter Text"),
        gr.Textbox(label="Source Language Code (e.g. en)", value="en"),
        gr.Textbox(label="Target Language Code (e.g. hi)", value="hi")
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="Language Translation Tool"
)

demo.launch()
```

```python
import gradio as gr
from googletrans import Translator

translator = Translator()

languages = {
    "English": "en",
    "Hindi": "hi",
    "French": "fr",
    "Spanish": "es",
    "German": "de",
    "Japanese": "ja",
    "Chinese": "zh-cn",
    "Arabic": "ar"
}

def translate_text(text, source_lang, target_lang):
    src_code = languages[source_lang]
    dest_code = languages[target_lang]
    result = translator.translate(text, src=src_code, dest=dest_code)
    return result.text

demo = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Enter Text", placeholder="Type text here..."),
        gr.Dropdown(choices=list(languages.keys()), label="Source Language", value="English"),
        gr.Dropdown(choices=list(languages.keys()), label="Target Language", value="Hindi")
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="🌍 Language Translation Tool",
    description="Enter text, select languages, and get instant translation!"
)

demo.launch()
```
